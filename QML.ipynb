```python
def circuit_expection(circuit,n_qubits,param_resolver): 
    simulator = cirq.Simulator() 
    result = simulator.simulate(circuit,param_resolver=param_resolver) 
    final_state = result.final_state_vector 
    exp_val = cirq.expectation_from_state_vector(
    state_vector=final_state,
    observables=cirq.Z(n_qubits[-1]),
    qid_shape=None
    )           
    
    return float(exp_val)
```

```python
def predict_value(circuit,n_qubits,params,param_vector): 
    param_resolver = {symbol: value for symbol, value in zip(params.values(), param_vector)}
    out_put = circuit_expection(circuit,n_qubits,param_resolver)     
    predict = (out_put+1)/2; 
    return predict  
```

```python
def cost_function(param_vector,circuit,n_qubits,params,X_train,Y_train,encode_input): 
    total_loss = 0 
    for i in range(len(X_train)): 
        sample = circuit.copy() 
        sample = encode_input(sample,X_train[i]) 
        predict = predict_value(sample,n_qubits,params,param_vector)
        actual= Y_train[i]
        total_loss += (predict-actual)**2
    return total_loss/len(X_train) 
```

```python
def train_quantum_model(circuit, qubits, params, X_train, Y_train, encode_input):
    param_vector = np.random.uniform(0, 2*np.pi, len(params))  
    result = minimize(
        cost_function,
        param_vector,
        args=(circuit, qubits, params, X_train, Y_train, encode_input),
        method="COBYLA",
        options={"maxiter": 80}
    )
    
    print("Final Loss:", result.fun)
    trained_params = result.x
    return trained_params 
```
